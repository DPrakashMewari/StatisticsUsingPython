* Central Limit Theoram : The central Limit theoram relies on the concept of sampling Distribution 
which is the probablity Distribution of statistic for large no of samples taken from population 
It says that sampling distribution of mean will always be normally Distributed . As long as the sample
size is large enough. Regardless of whether the population has a normal,poisson,Binomial or Any other 
Distribution The sampling Distribution of the Mean will be Normal .

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# set parameters
population_mean = 50
population_std = 10
sample_size = 1000
num_samples = 1000

# generate population data
population_data = np.random.normal(population_mean, population_std, 100000)

# generate sample means
sample_means = []
for i in range(num_samples):
    sample = np.random.choice(population_data, sample_size)
    sample_means.append(np.mean(sample))

# plot the sample means
plt.hist(sample_means, bins=30, density=True)

# calculate mean and standard deviation of sample means
sample_mean = np.mean(sample_means)
sample_std = np.std(sample_means)

# plot the normal distribution
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
y = stats.norm.pdf(x, sample_mean, sample_std)
plt.plot(x, y)

plt.title("Central Limit Theorem Demonstration")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")
plt.show()
```

* Hypothesis Testing : Hypothesis testing is a statistical method used to make decisions about a population parameter based on a sample of data. It involves setting up two mutually exclusive hypotheses, the null hypothesis (H0) and the alternative hypothesis (Ha).
        
Null Hypothesis is Existing Belief and Alternate Hypothesis is what we intend to establish with 
New Evidence 
        
!Example : Suppose a company wants to determine whether a new advertising campaign has increased sales of their product. The null hypothesis (H0) would be that there is no difference in sales between the pre-advertising period and the post-advertising period, while the alternative hypothesis (Ha) would be that there is an increase in sales.

Steps For hypothesis test are as follows: 
    
1. Define Null and Alternate Hypothesis(From Population Parameter )
  [ A. Ho Null Hypothesis ,B. H1 Alternate Hypothesis ]
2. Identify the test Statistic to be used for testing the validity of Null Hypothesis 
     Example z-test ,T-test 
    
3. Decide the criteria for rejection and retention of Null Hypothesis . This is called Significance 
   value alpha . Mostly used 0.5

4. Calculate the P value which is the conditional Probablity of observing the test statistic value 
    When Null Hypothesis is true 

5. Take the Decison to reject the Null Hypothesis or retain it based on P-value and the significance 
value alpha

* Z-test : The Z-test is a statistical hypothesis test used 
    to determine whether two population means are different,
    given a known population variance and sample size.
    It is a parametric test that assumes that the data follow a normal distribution.
    
 1) When to Use this : 
    The Z-test is used when:
       - The sample size is large (usually n > 30).
       - The population standard deviation (σ) is known.
       - The data follows a normal distribution.

** The Z-test compares the difference between two means (usually a sample mean and a population mean) to the standard error of the difference between the means. If the difference between the means is greater than what would be expected by chance alone (i.e., it is statistically significant), then the null hypothesis (that the means are equal) is rejected.

```python
# Example-1
# One-Sample Z-Test:
# Suppose we want to test if the mean weight of a population of apples is 150g. 
# We have a sample of 25 apples with a mean weight of 155g and a population 
# standard deviation of 5g. We want to test if the sample mean is significantly different
# from the population mean using a significance level of 0.05.
```

```python
import numpy as np
from statsmodels.stats.weightstats import ztest

# Sample data
sample = np.random.normal(loc=155, scale=5, size=25)

# One-sample Z-test
# Ztest takes population mean and sample mean 
z_score, p_value = ztest(x1=sample, value=150)

print('Z-score:', z_score)
print('p-value:', p_value)

if p_value < 0.05:
    print('Reject null hypothesis')
else:
    print('Fail to reject null hypothesis')
```

```python
# Two-Sample Z-Test:

# Suppose we want to test if there is a 
# significant difference between the mean weight of two populations of apples.
# We have a sample of 25 apples from population 1 with a mean weight of 150g and
# a standard deviation of 
# 5g, and a sample of 30 apples from population 2 with a mean weight of 155g and 
# a standard deviation of 6g. 
# We want to test
# if the two sample means are significantly different using a significance level of 0.05
```

```python
# Sample data
# scale is stndard deviation
sample1 = np.random.normal(loc=150, scale=5, size=25)
sample2 = np.random.normal(loc=155, scale=6, size=30)

# Two-sample Z-test
z_score, p_value = ztest(x1=sample1, x2=sample2, alternative='two-sided')

print('Z-score:', z_score)
print('p-value:', p_value)

if p_value < 0.05:
    print('Reject null hypothesis')
else:
    print('Fail to reject null hypothesis')
```

2. T-test : The t-test is a statistical hypothesis test used to determine if there is a significant difference between the means of two groups. The t-test is used when the population 
standard deviation is unknown and the sample size is small (less than 30).

Example :  suppose we want to know if there is a significant difference in the heights of boys and girls in a particular school. We take a sample of 20 boys and 20 girls and measure their heights in centimeters. We can then use a t-test to determine if the difference in the mean heights between the two groups is significant.

Here are the steps to perform a t-test:

- State the null and alternative hypotheses. In our example, the null hypothesis would be that there is no significant difference in the mean heights of boys and girls, while the alternative hypothesis would be that there is a significant difference.

- Calculate the t-statistic. This involves calculating the difference between the means of the two groups and dividing it by the standard error of the difference.

- Determine the degrees of freedom. This is equal to the sample size minus one.

- Determine the critical value of t. This is based on the significance level (alpha), the degrees of freedom, and whether the test is one-tailed or two-tailed.

- Compare the calculated t-statistic to the critical value of t. If the calculated t-statistic is greater than the critical value, we reject the null hypothesis and conclude that there is a significant difference between the means of the two groups. Otherwise, we fail to reject the null hypothesis.

```python
# One-sample t-test
# Suppose we have a sample of 10 observations and we want to test if the mean 
# of the population is equal to 5.

import numpy as np
from scipy.stats import ttest_1samp

# Create a sample of 10 observations
sample = np.array([4, 6, 5, 3, 7, 6, 5, 4, 5, 6])
# print(np.mean(sample))

# Perform a one-sample t-test
t_statistic, p_value = ttest_1samp(sample, 5)

print("t-statistic:", t_statistic)
print("p-value:", p_value)


#  It indicates that we fail to reject the null hypothesis that the population mean is equal to 5.
```

```python
# Two sample T-test 
# The null hypothesis for a two-sample t-test is that there is no difference between 
# the means of the two groups. 
# The alternative hypothesis is that there is a significant difference.


import numpy as np
from scipy.stats import ttest_ind

# Generate two samples of data
sample1 = np.array([7.8, 6.5, 8.2, 5.6, 6.9, 7.4, 8.1, 7.6, 7.9, 6.8])
sample2 = np.array([5.5, 6.1, 5.8, 6.7, 5.4, 7.2, 6.4, 6.0, 5.7, 6.3])

# Perform the two-sample t-test
t_statistic, p_value = ttest_ind(sample1, sample2)

# Print the results
print("t-statistic:", t_statistic)
print("p-value:", p_value)

# Since the p-value is less than the significance level (0.05),
# we reject the null hypothesis and conclude that there is
# a significant difference between the means of the two groups.
```

* Chisquare Test : The Chi-Square test is a statistical test that is used to determine whether there is a significant difference between an observed frequency distribution and an expected frequency distribution. It is based on the Chi-Square distribution and is used to test hypotheses about categorical data  

1. Expected Data from population Data
2. Observed Data from Sample Data 

Then Make Hypothesis who is H0 and H1

Goodness of Fit:
Suppose a researcher is studying the distribution of eye color in a population. The researcher believes that 60% of the population has brown eyes, 20% has blue eyes, 10% has green eyes, and 10% has other eye colors. The researcher takes a random sample of 100 individuals and records their eye color. The observed frequencies of eye color in the sample are as follows: 70 brown, 15 blue, 8 green, and 7 other. The researcher can use the Chi-Square test to determine if the observed frequencies are significantly different from the expected frequencies.

! Chi-Square test can be used to test whether the observed data deviates significantly from what we would expect if there were no relationship between the two variables. If the test statistic is large enough, we can reject the null hypothesis of no relationship and conclude that there is evidence of a significant relationship between the variables.

```python
import numpy as np
from scipy.stats import chi2_contingency

# Define the contingency table with sample data
data = np.array([[15, 10, 5], [10, 20, 10], [5, 10, 15]])

# Perform the Chi-Square test
chi2, pval, dof, expected = chi2_contingency(data)

# Print the results
print(f"Chi-Square value: {chi2}")
print(f"P-value: {pval}")
print(f"Degrees of freedom: {dof}")
print("Expected frequencies:")
print(expected)


#1. # The Chi-Square value measures the difference between 
# the observed and expected frequencies, with larger values indicating a larger difference


#2. The p-value tells us the probability of observing a Chi-Square value
# at least as extreme as the one we obtained

# Conclusion
# If the p-value is less than our chosen level of significance (e.g., 0.05), 
# we can reject the null hypothesis and conclude that there is
# a significant difference between the observed and expected frequencies.
```

```python
# Chi-Square Test for Population Data

import numpy as np
from scipy import stats

# Define the observed and expected frequencies
observed = np.array([40, 30, 20, 10])
expected = np.array([25, 25, 25, 25])

# Perform the Chi-Square test
chi2, pval = stats.chisquare(observed, f_exp=expected)

# Print the results
print(f"Chi-Square value: {chi2}")
print(f"P-value: {pval}")

# The Chi-Square value measures the difference between the observed and expected frequencies,
# with larger values indicating a larger difference

# The p-value tells us the probability of observing a Chi-Square value at 
# least as extreme as the one we obtained


# If the p-value is less than our chosen level of significance (e.g., 0.05), 
# we can reject the null hypothesis and conclude that there is a
# significant difference between the observed and expected frequencies.
```

```python
# Chisquare Distribution 
import numpy as np
from scipy.stats import chi2
import matplotlib.pyplot as plt

# Define the degrees of freedom
df = 5   # as we increase degree of freedom it become Normal Distribution

# Generate some random numbers from a Chi-Square distribution
x = np.random.chisquare(df, size=1000)

# Calculate the PDF values for a range of x values
x_values = np.linspace(0, 20, 1000)
pdf_values = chi2.pdf(x_values, df)

# Plot the PDF of the Chi-Square distribution
fig, ax = plt.subplots()
ax.plot(x_values, pdf_values, label=f"df={df}")
ax.hist(x, density=True, bins=20, alpha=0.5)
ax.legend()
ax.set_xlabel("x")
ax.set_ylabel("PDF")
ax.set_title("Chi-Square Distribution")
plt.show()
```
