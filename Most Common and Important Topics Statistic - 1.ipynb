### Under Descriptive : Measure of Central Tendency
- Mean : It actually represents the average of the given collection of data. It is applicable for both continuous and discrete data.
- Median : It represents the mid-value of the given set of data when arranged in a particular order.
- Mode : The Most frequent number occuring in the data is know as the mode.


```python
# let Do this Python :

# Mean
n1 = [1,10,11,2,4]
n_num = [1, 2, 3, 4, 5]
def mean_find(n1):
    mean = sum(n1)/len(n1)
    print("Your Mean is ",mean)
mean_find(n1)
mean_find(n_num)
```

```python
# Median:

def median_find(n1):
    n = len(n1)
    n1.sort()
    if n %2 == 0:
        m1 = n1[n//2]
        m2 = n1[n//2-1]
        median = (m1+ m2)/2
    else:
        median = n1[n//2]
    print("Your median is ",median)

n = n_num.copy()
print(n)
median_find(n)
```

```python
# Mode:

from collections import Counter


def mode_find(n1):    
    data = Counter(n1)
    n = len(n1)
    mode = [x for x,y in data.items() if y == max(list(data.values()))] 
    if len(mode) == n: 
        pass
    print("Mode is ",",".join(map(str,mode)))


n1 = n_num.copy()
n1.append(5)
print(n1)
y= [11, 8, 8, 3, 4, 4, 5, 6, 6, 6, 7, 8]
mode_find(y)
```

#### Skewness is a stastical term and it is way to estimate or measure the shape of a distribution
    1. Symmetrical: A distribution can be called symmetric if it appears the same from the left and right from the center point.
    2. Asymmetrical: A distribution can be called asymmetric if it doesn’t appear the same from the left and right from the center point.

### Kurtosis : It determines whether a distribution is heavy-tailed in respect of the normal distribution.

- kurtosis for normal distribution is equal to 3.
- For a distribution having kurtosis < 3: It is called playkurtic.
- For a distribution having kurtosis > 3, It is called leptokurtic and it signifies that it tries to produce more outliers rather than the normal distribution.







```python
# Let do using python - Skewness 
from scipy.stats import skew
# Creating a dataset
dataset = [10, 25, 14, 26, 35, 45, 67, 90,40, 50, 60, 10, 16, 18, 20]

print(skew(dataset, axis=0, bias=True))

plt.hist(dataset) # We can see it left skew and skewness value also shown it is left skew

```

#### Covariance : Covariance is a single number we can calculate from a list of paired values.
    It tells us if the paired values tend to rise together, or if one tends to rise as the other falls.

- positive says they rise and fall together
- negative says that one rises as the other falls

1. If COV(xi, xj) = 0 then variables are uncorrelated
2. If COV(xi, xj) > 0 then variables positively correlated
3. If COV(xi, xj) > < 0 then variables negatively correlated

```python
# Find Covariance : 
x = [1.23, 2.12, 3.34, 4.5]
 
y = [2.56, 2.89, 3.76, 3.95]
 
cov_mat = np.stack((x, y), axis = 0)
 
print(cov_mat)
print(np.cov(cov_mat))

# plt.scatter(x=x,y=y)
```

#### Correlation: When two sets of data are strongly linked together we say they have a High Correlation.

1. Correlation is Positive when the values increase together, and
2. Correlation is Negative when one value decreases as the other increases




Correlation can have a value:

- 1 is a perfect positive correlation
- 0 is no correlation (the values don't seem linked at all)
- -1 is a perfect negative correlation

```python
# Lets find 
# calculate the Pearson's correlation between two variables
from numpy.random import randn
from scipy.stats import pearsonr

# prepare data
data1 = 20 * randn(1000) + 100
data2 = data1 + (10 * randn(1000) + 50)

# calculate Pearson's correlation
corr, _ = pearsonr(data1, data2)
print('Pearsons correlation: %.3f' % corr)


plt.scatter(x=data1,y=data2)  # High Positive Correlation
```

```python
# Let do Using PDF  
from scipy.stats import norm

#Creating a series of data of in range of 1-50.
data = np.linspace(1,50,200)

#Creating a Function . # Normal Distribution PDF (We can verify using formula)
def normal_dist(data, mean ,stddev):
    prob_density =(np.pi*stddev)* np.exp(-0.5*((data-mean) /stddev) **2)
    return prob_density/100

#Calculate mean and Standard deviation .
mean = np.mean(data)
stddev = np.std(data)

#Apply function to the data.
pdf = normal_dist(data, mean, stddev)

#Plotting the Results
plt.plot (data, pdf)
plt.xlabel ( 'Data points ')
plt.ylabel ( "Probability Density ")
plt.show()
```

* lognormal distribution:- is a continuous probability distribution of a random variable whose logarithm is normally distributed. Thus, if the random variable X is log-normally distributed, then Y = ln(X) has a normal distribution.

```python
from scipy.stats import lognorm
mu, sigma = 3., 1. # mean and standard deviation
s = np.random.lognormal(mu, sigma, 1000)
count, bins, ignored = plt.hist(s, 100, density=True, align='mid')
x = np.linspace(min(bins), max(bins), 10000)
pdf = (np.exp(-(np.log(x) - mu)**2 / (2 * sigma**2))
       / (x * sigma * np.sqrt(2 * np.pi)))  # Formula using 
plt.plot(x, pdf)
```

* Power Law Distribution : Power law is a functional relationship between two quantities, 
    where a relative change in one quantity results in a proportional relative
    change in the other quantity  (80-20 Rule)


!Example : 80% of sale comming from 20% of overall Product !<br>
!Example : 80% of the window crashes coming from 20% of bugs !

```python
import numpy as np
from scipy.stats import powerlaw
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1, 1)
a = 0.659 # 65 percent confidence Interval
mean, var, skew, kurt = powerlaw.stats(a, moments='mvsk')  # To Plot a data we need mean,var,skew,kurt
x = np.linspace(powerlaw.ppf(0.01, a),
                powerlaw.ppf(0.99, a), 100)
ax.plot(x, powerlaw.pdf(x, a),
       'r-', lw=5, alpha=0.6, label='powerlaw pdf')
```

* Bernaulli Distribution : The Bernoulli distribution is a probability distribution 
                             that describes a single
                             experiment that can have one of two outcomes - success or failure


!where the outcome of an experiment is either 0 or 1. For example, flipping a coin can be modeled using a Bernoulli distribution, where success is defined as getting heads (1) and failure is defined as getting tails (0).


```python
import scipy.stats as stats
import numpy as np
import matplotlib.pyplot as plt

# Define the probability of success
p = 0.6

# Create a Bernoulli distribution object
bernoulli_dist = stats.bernoulli(p)

# Generate 1000 random numbers from the distribution
random_numbers = bernoulli_dist.rvs(size=1000)

# Plot a histogram of the data
plt.hist(random_numbers, bins=[-0.5, 0.5, 1.5], rwidth=0.8)

# Add labels and title to the plot
plt.xlabel('Outcome')
plt.ylabel('Count')
plt.title('Bernoulli Distribution')

# Show the plot
plt.show()
```

* Poisson Distribution : The Poisson distribution is a discrete probability distribution
that is used to model the probability of a 
certain number of rare and random events occurring in a fixed interval of time or space


!Suppose you are managing a call center, and you want to model the probability of receiving a certain number of calls in a given hour. You have historical data that tells you that on average, you receive 3 calls per hour. You can use the Poisson distribution to model the probability of receiving a certain number of calls in a given hour. 

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Define the average rate of occurrence
lambda_ = 3

# Create a Poisson distribution object
poisson_dist = poisson(mu=lambda_)

# Generate 100 random numbers from the distribution
random_numbers = poisson_dist.rvs(size=100)

# Plot the data
plt.hist(random_numbers)
plt.title("Poisson Distribution")
plt.xlabel("Number of Events")
plt.ylabel("Frequency")
plt.show()
```

* Binomial Distribution : The binomial distribution is a discrete probability distribution that is       used to model the probability of a certain number of successes in a fixed number of independent trials
            
!Example : Suppose you are a salesperson, and you know that the probability of making a sale to a potential customer is 0.2. You have 10 potential customers to contact today. You can use the binomial distribution to model the probability of making a certain number of sales. For example, the probability of making 2 sales today is:-
- Formula_P(X=k) = (n choose k) * p^k * (1-p)^(n-k)

P(X=2) = (10 choose 2) * 0.2^2 * 0.8^8 = 0.302

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Define the number of trials
n = 10

# Define the probability of success
p = 0.5

# Create a binomial distribution object
binomial_dist = binom(n, p)

# Generate 100 random numbers from the distribution
random_numbers = binomial_dist.rvs(size=100)

# Plot the data
plt.hist(random_numbers)
plt.title("Binomial Distribution")
plt.xlabel("Number of Successes")
plt.ylabel("Frequency")
plt.show()
```

* Uniform Distribution : A continuous uniform distribution is defined over a continuous interval,
    and has a constant probability density function (pdf) within that interval. 
    The pdf is zero outside the interval. 
    An example of a continuous uniform distribution is the distribution of 
    heights of adult humans between a certain minimum and maximum value.
<br>    <br>
* Discrete uniform distribution : A discrete uniform distribution is defined over a finite set of outcomes, and assigns equal probabilities to each outcome. An example of a discrete uniform distribution is the roll of a fair die, where each of the six possible outcomes has a probability of 1/6.

```python
# Continous Uniform Distribution : 
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform

# Define the minimum and maximum values of the interval
a = 0
b = 20

# Create a uniform distribution object
uniform_dist = uniform(loc=a, scale=b-a)

# Generate 100 random numbers from the distribution
random_numbers = uniform_dist.rvs(size=100)

# Plot the probability density function
x = np.linspace(a, b, 100)
pdf = uniform_dist.pdf(x)
plt.plot(x, pdf)
plt.title("Continuous Uniform Distribution")
plt.xlabel("X")
plt.ylabel("Probability Density")
plt.show()
```
